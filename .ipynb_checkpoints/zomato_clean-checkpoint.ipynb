```python
import pandas as pd
import numpy as np
import re
dataset=pd.read_csv('C:\\Users\\Dell\\zomato.csv')
dataset.head()
```

```python
dataset.info()
```

```python
null_var=dataset.isnull().sum()/dataset.shape[0] *100
null_var
```

```python
dataset=dataset.drop(columns=["url","address","phone","reviews_list","listed_in(city)","dish_liked","menu_item"])
dataset
```

```python
dataset.rename({"name":"Restaurants",
           "rate":"Rating",
           "approx_cost(for two people)":"Cost",
           "listed_in(type)":"type",},inplace=True,axis=1)
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.shape 
```

```python
dataset["Rating"].unique()
```

```python
def rating(rate):
    if rate=='-'or rate=='NEW':
        return np.nan
    else:
        rate =str(rate).split('/')[0]
        return float(rate)
    
dataset["Rating"] = dataset["Rating"].apply(rating)
dataset["Rating"].head(5)
```

```python
dataset
```

```python
dataset["Cost"].unique()
```

```python
def costing(cost):
    cost=str(cost)
    if "," in cost:
        cost=cost.replace(",","")
        return float(cost)
    else:
        return float(cost) 
        
dataset["Cost"] = dataset["Cost"].apply(costing)
dataset["Cost"].head(5)
```

```python
dataset['Restaurants'] = dataset['Restaurants'].str.replace('[Ãx][^A-Za-z]+','',regex=True)
```

```python
dataset["Rating"] = dataset["Rating"].fillna(0) 
```

```python
dataset.dropna(inplace=True)
```
